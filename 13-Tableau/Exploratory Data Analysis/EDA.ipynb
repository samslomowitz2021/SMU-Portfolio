```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file = "2018b_citybike_rides_sample.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df2 = df.dropna()
```

```python
df2
```

```python
df2.info()
```

```python
df2.gender.value_counts()
```

```python
df2["birth year"].value_counts(ascending=False)
```

```python
df2["birth year"].nunique()
```

```python
sns.histplot(data=df2, x="birth year", color = "#58595B")
plt.show()
```

```python
mask = (df2["birth year"] > 1940) 
df3 = df2.loc[mask,:]
df3
```

```python
df3["birth year"].value_counts(ascending=False)
```

```python
df3.describe()
```

```python
df_plot = df3.groupby("gender").tripduration.mean().reset_index().sort_values(by="tripduration", ascending=False)

# then plot
plt.figure(figsize=(10,5))

plt.bar(df_plot.gender, df_plot.tripduration, color="seagreen")

plt.xticks(rotation=45)
plt.title("Avg Trip Duration by Gender")
plt.ylabel("Trip Duration (s)")
plt.grid(axis="y",alpha=0.15)

plt.show()
```

```python
df_plot
```

```python


df_plot = df3.groupby("usertype").tripduration.mean().reset_index().sort_values(by="tripduration", ascending=False)

# then plot
plt.figure(figsize=(10,5))

plt.bar(df_plot.usertype, df_plot.tripduration, color="seagreen")

plt.xticks(rotation=45)
plt.title("Avg Trip Duration by User Type")
plt.ylabel("Trip Duration (s)")
plt.grid(axis="y",alpha=0.15)

plt.show()
```
