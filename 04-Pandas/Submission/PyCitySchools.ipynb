```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
schools_count = school_data_complete.school_name.nunique()
schools_count
```

```python
students_count = school_data_complete["Student ID"].nunique()
students_count
```

```python
#https://stackoverflow.com/questions/32072076/find-the-unique-values-in-a-column-and-then-sort-them
a = school_data_complete['budget'].drop_duplicates()
b = a.sum()
```

```python
math = school_data_complete.math_score.mean()
math
```

```python
reading = school_data_complete.reading_score.mean()
reading
```

```python
passed_math = school_data_complete.loc[school_data_complete["math_score"]>69.9999999,:] 
passed_math2 = passed_math["math_score"].count() 
passed_math3 = passed_math2 / students_count * 100
passed_math3 
```

```python
passed_reading = school_data_complete.loc[school_data_complete["reading_score"]>69.9999999,:] 
passed_reading2 = passed_reading["reading_score"].count() 
passed_reading3 = passed_reading2 / students_count * 100
```

```python
# passed_both = school_data_complete.loc[(school_data_complete["math_score"]>69.9999999) | (school_data_complete["reading_score"]>69.9999999),:]
# passed_both["reading_score"].count()

school_data_complete2 = school_data_complete[['math_score','reading_score']][(school_data_complete["math_score"] >69.9999999) & (school_data_complete["reading_score"] >69.9999999)]

school_data_complete3 = school_data_complete2.count()

school_data_complete4 = school_data_complete3 / students_count 
school_data_complete5 = school_data_complete4["math_score"] *100


```

```python


series_1 = pd.Series([schools_count])
series_2 = pd.Series([students_count])
series_3 = pd.Series([b])
series_4 = pd.Series([math])
series_5 = pd.Series([reading])
series_6 = pd.Series([passed_math3])
series_7 = pd.Series([passed_reading3])
series_8 = pd.Series([school_data_complete5])

df = pd.DataFrame(columns = ['Total Schools'])
df = pd.DataFrame(columns = ['Total Students'])
df = pd.DataFrame(columns = ['Total Budget'])
df = pd.DataFrame(columns = ['Average Math Score'])
df = pd.DataFrame(columns = ['Average Reading Score'])
df = pd.DataFrame(columns = ['% Passing Math'])
df = pd.DataFrame(columns = ['% Passing Reading'])
df = pd.DataFrame(columns = ['% Overall Passing'])


df['Total Schools'] = series_1
df['Total Students'] = series_2
df['Total Budget'] = series_3
df['Average Math Score'] = series_4
df['Average Reading Score'] = series_5
df['% Passing Math'] = series_6
df['% Passing Reading'] = series_7
df['% Overall Passing'] = series_8

df2 = df[['Total Schools', 'Total Students', 'Total Budget',
       'Average Math Score', 'Average Reading Score', '% Passing Math',
       '% Passing Reading', '% Overall Passing']]
df2
df2.style.format({'Total Schools':"{:20,.0f}", 'Total Students':"{:20,.0f}", 'Total Budget':"{:20,.0f}",
       'Average Math Score':"{:20,.2f}", 'Average Reading Score':"{:20,.2f}", '% Passing Math':"{:20,.2f}",
       '% Passing Reading':"{:20,.2f}", '% Overall Passing': "{:20,.2f}"})
    
              
```

```python
#https://pandas.pydata.org/docs/reference/api/pandas.core.groupby.DataFrameGroupBy.aggregate.html

school_type = school_data_complete.groupby(["school_name"]).agg({"type": "max", "Student ID": "count", "budget": "max" })
school_type
```

```python
school_type["per Student Budget"] = school_type.budget / school_type["Student ID"]
school_type300 = school_type
school_type300
```

```python
#https://www.kite.com/python/answers/how-to-drop-the-index-column-of-a-pandas-dataframe-in-python#:~:text=reset_index()%20to%20drop%20the,reset%20the%20index%20of%20pandas.

school_type2 = school_data_complete.groupby(["school_name"]).agg({"math_score":"mean"})
school_type2
```

```python
school_type3 = school_data_complete.groupby(["school_name"]).agg({"reading_score":"mean"})
school_type3
school_type3
```

```python
# school_data_complete['% Passed math_score'] = (school_data_complete['math_score']>69.99999999)

# passed_math20 = school_data_complete['% Passed math_score'].astype(int)
# passed_math20.sum() / len(passed_math20)

# school_data_complete

school_data_complete['% Passed math_score2']=school_data_complete['math_score'].apply(lambda x: 1 if x >69.99999 else 0)
math500 = school_data_complete.groupby(["school_name"])["% Passed math_score2"].mean() *100
math500
```

```python
school_data_complete['% Passed reading_score2']=school_data_complete['reading_score'].apply(lambda x: 1 if x >69.99999 else 0)
reading500 = school_data_complete.groupby(["school_name"])["% Passed reading_score2"].mean() *100
reading500
```

```python
school_data_complete['% Passed both_score2']=(school_data_complete['math_score']>69.999999) & (school_data_complete["reading_score"] > 69.99999)
result = school_data_complete['% Passed both_score2'].astype(int)


school_data_complete["% Passed both_score3"] = result
school_data_complete

both500 = school_data_complete.groupby(["school_name"])["% Passed both_score3"].mean() *100
both500
```
